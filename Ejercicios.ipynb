```python
import manim as mn
from manim import *
config.media_width = "75%"
config.verbosity = "WARNING"
print(mn.__version__)
```

```python
%%manim -ql MiEscena
class MiEscena(Scene):
    def construct(self):
        circulo=Circle()
        circulo.set_fill(PINK, opacity=1)
        self.play(Create(circulo))
        
```

```python
%%manim -ql MiEscena2
class MiEscena2(Scene):
    def construct(self):
        cuadrado=Square()
        cuadrado.rotate(PI/4).scale(3)
        circulo=Circle()
        circulo.set_fill(RED,opacity=0.8).scale(2)
        self.play(Create(cuadrado))
        self.wait(0.5)
        self.play(Transform(cuadrado,circulo))
        self.wait(0.5)
        self.play(FadeOut(cuadrado))
```

```python
%%manim -ql MiEscena3
class MiEscena3(Scene):
    def construct(self):
        circulo=Circle()
        circulo.set_fill(RED,opacity=0.6)
        cuadrado=Square()
        cuadrado.set_fill(BLUE,opacity=0.5)
        cuadrado.next_to(circulo, LEFT, buff=1)
        self.play(Create(circulo),Create(cuadrado))
        self.wait()
```
